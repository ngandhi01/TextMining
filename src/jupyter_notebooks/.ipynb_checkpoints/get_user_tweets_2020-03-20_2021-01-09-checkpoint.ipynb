### Get tweets from Twitter

Now that we have these user IDs, let's use Tweepy to interface with the Twitter API in order to get all the tweets from these users. 

```python
def authenticate(consumer_key, consumer_secret, access_token, access_secret):

    """
      Allows authentication with Twitter API, with relevant IDs
      Input: IDs
      Output: authentication, API access
    """

    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_secret)

    api = tweepy.API(auth)

    try:
        api.verify_credentials()
        print("Authentication OK")
    except Exception as e:
        print("Error during authentication")
        print(e)

    return auth, api
```

```python
# get authentication
consumer_key = ''
consumer_secret = ''
access_key = ''
access_secret = ''

with open("../../twitter_creds.txt", 'r') as twitter_creds:
    consumer_key = twitter_creds.readline().rstrip() # reads line, removes trailing whitespaces
    consumer_secret = twitter_creds.readline().rstrip()
    access_key = twitter_creds.readline().rstrip()
    access_secret = twitter_creds.readline().rstrip()

try:
    auth, api = authenticate(consumer_key, consumer_secret, access_key, access_secret)
except Exception as e:
    print("Authentication failed")
    print(e)
```

Checking out the Twitter API: https://developer.twitter.com/en/docs/twitter-api
Description of what the fields are: https://developer.twitter.com/en/docs/twitter-api/v1/data-dictionary/object-model/user

What fields might be good to keep track of?

1. favourites_count
2. followers_count
3. friends_count
4. id_str (this is the user ID - their docs say to use this string rather than the int id)
5. statuses_count (number of posts + retweets)
6. created_at (indicates when account was created. Might be interesting to see if some people made accounts during COVID times and if they've reacted differently than others who have had accounts for a while)



```python
def get_user_info(user_id):
    
    """
        Takes user ID, makes call to Twitter API, and gets relevant info about the user profile
    """
    
    user = api.get_user(user_id)._json
    
    id_str = user["id_str"]
    created_at = user["created_at"]
    statuses_count = user["statuses_count"]
    favourites_count = user["favourites_count"]
    followers_count = user["followers_count"]
    friends_count = user["friends_count"]
    
    return [id_str, created_at, statuses_count, favourites_count, followers_count, friends_count]
    
    
```

### Clean tweets

Now that we have all the tweet text for each user, let's clean them all up

```python
PUNCTUATION ='''!()-[]{};:'"\,<>./?@$%^&*_~''' # keep hashtags
STOPWORDS = stopwords.words("english")
```

```python
def remove_emoji(string):
    """
        Removes emojis
    """
    text = string.encode("utf-8")
    allchars = [str for str in text.decode('utf-8')]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.decode('utf-8').split() if not any(i in str for i in emoji_list)])
    return clean_text
```

```python
def clean_text(text, include_hashtag_symbol = True):
    """
        Removes punctuation, does string split, and removes links
        
        Works on individual tweets
        
        Turns tweet string into tokenized array of cleaned text
        
        Allows you to keep hashtags as they are or remove the hashtags
    """
    
    return_arr = []
    
    # remove punctuation
    text_no_punctuation = ""
    
    for char in text:
        if char not in PUNCTUATION:
            text_no_punctuation = text_no_punctuation + char
            
    # remove emojis
    text_no_punctuation = remove_emoji(text_no_punctuation)
    text_no_punctuation = re.sub(r'\\U[a-zA-Z0-9]{8}', '', text_no_punctuation)
    
    # remove \n and \t
    text_no_punctuation = re.sub(r'\n', '', text_no_punctuation)
    text_no_punctuation = re.sub(r'\t', '', text_no_punctuation)
    
    # remove escape sequences
    text_no_escape = ""
    
    for char in text_no_punctuation:
        try:
            char.encode('ascii')
            text_no_escape = text_no_escape + char # this'll catch chars that don't have an ascii equivalent (e.g., emojis)
        except:
            pass
    
    # add space between # and another char before it (e.g., split yes#baseball into yes #baseball)
    text_no_escape = re.sub(r"([a-zA-Z0-9]){1}#", r"\1 #", text_no_escape)
    
    # other preprocessing
    text_arr = text_no_escape.split(' ')
    
    for word in text_arr:
        
        # clean words
        word = word.lower()
        
        if "http" not in word and word.strip() != '' and word not in STOPWORDS:
            
            # optional - remove hashtag from words
            if not include_hashtag_symbol:
                
                if "#" in word:
                    new_word = ""
                    for char in word:
                        if char != '#':
                            new_word = new_word + char
                    return_arr.append(new_word)
                    
                else:
                    return_arr.append(word)
            
            else:
                # otherwise, return word as is
                return_arr.append(word)
            
    return return_arr
```

```python
tweets_compiled["cleaned_text"] = tweets_compiled["tweet_text"].apply(lambda x : clean_text(x))
tweets_compiled["cleaned_text_no_hashtags"] = tweets_compiled["tweet_text"].apply(lambda x : clean_text(x, False))
```
