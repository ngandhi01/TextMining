### Set up our own requests

Our goal is to get (1) user info and (2) past user tweets. 

So, for each user, we want to get their information and their past tweets

Example for setting up query string:
https://developer.twitter.com/en/docs/twitter-api/premium/rules-and-filtering/using-premium-operators

```python
def get_past_tweets(bearer_token, user_id=None, search_terms=None, start_date="20200101", end_date="20201231", max_results=100):
    
    """
        Uses Twitter Premium API, with Historical Search Full Archive endpoint, to get past tweets
        
        Args:
            • search_terms: list of terms to look up (type:list)
            • start_date: start date, in the format YYYYMMDD (type:str)
            • end_date: end_date, in the format YYYYMMDD (type:str)
            • bearer_token: long string containing bearer token creds for application (type:str)
            • max_results: maximum # of results to return (default: 100) (type: int)
            • user_id: id of user to look up (type:int or str)
            
        Returns list of tweet objects. Each item in the list is a tweet object, and the 
        text of the tweet can be accessed using the "text" key. 
        
        e.g., 
            for tweet in tweet_list:
                print(tweet["text"])
    """
    
    if type(start_date) != str or type(end_date) != str:
        raise Exception("The start and end dates need to be strings in the format YYYYMMDD")
    
    if len(start_date) != 8 or len(end_date) != 8:
        raise Exception("The start and end dates have the wrong length. They must be in the format YYYYMMDD")
        
    if user_id is None:
        raise Exception("Need to look up a specific user")
        
    # set up params for request
    endpoint = "https://api.twitter.com/1.1/tweets/search/fullarchive/dev.json"
    headers = {"Authorization":f"Bearer {bearer_token}", "Content-Type": "application/json"}
    
    """
    if len(search_terms) > 0:
        query_str = search_terms[0]

        for idx in range(1, len(search_terms)):
            query_str += f" OR {search_terms[idx]}"
    """
    
    query_str = f"from:{user_id}"
    start = start_date + "0000"
    end = end_date + "0000"
        
    data_query = '{' + '"query":"(' + query_str + ')", "fromDate": "' + start + '", "toDate": "' + end + '", "maxResults":"' + str(max_results) + '"}'

    # send request
    response = requests.post(endpoint,data=data_query,headers=headers).json()
    
    # print results, for ease of viewing
    #print(json.dumps(response, indent = 2))
    
    # return results
    return response
```

### Scaling up

Now that we know that this works for one user, let's run this for a sample of 1,000 users and see if there's any possible signal, before doing it on the 250,000+ users.

Relevant rate limits: https://developer.twitter.com/en/docs/twitter-api/premium/rate-limits



```python
searched_user_ids_list = []
searched_tweet_text_list = []
searched_tweet_createdAt_list = []
searched_tweet_favoriteCount_list = []
searched_tweet_retweetCount_list = []
```

```python
for idx, user_id in enumerate(unique_user_ids[0:1000]):
    
    if idx % 50 == 0:
        print(f"Working on user # {idx} out of 1000")
        
    try:
        new_tweets = get_past_tweets(bearer_token, user_id=user_id, start_date="20200320")
    
        if "error" in new_tweets.keys():
            raise Exception(f"Error in getting previous tweets for user {user_id}: {new_tweets['error']}")
            
        for tweet in new_tweets["results"]:

            searched_user_ids_list.append(tweet["user"]["id_str"])
            searched_tweet_text_list.append(tweet["text"])
            searched_tweet_createdAt_list.append(tweet["created_at"])
            searched_tweet_favoriteCount_list.append(tweet["favorite_count"])
            searched_tweet_retweetCount_list.append(tweet["retweet_count"])
            
    except Exception as e:
        print(f"Error for user {user_id} at index = {idx}: {e}")
        willContinue = input("Would you like to continue? Select 'y' to continue. Select 'w' to wait. Press anything else to break")
        if willContinue == 'y':
            continue
        elif willContinue == 'w':
            getInput = True
            while getInput:
                try:
                    num_seconds_wait = int(input("How many seconds would you like to wait?"))
                    print(f"Ok, will wait {num_seconds_wait} seconds ({num_seconds_wait / 60} minutes)")
                    time.sleep(num_seconds_wait)
                    getInput = False
                    print(f"Now continuing loop, after waiting for {num_seconds_wait} seconds ({num_seconds_wait/60} minutes)")
                    continue
                except:
                    print("Need a valid number of seconds to wait.")   
        else:
            print("Exiting loop")
            break
            
    
```

```python
searched_user_tweets = pd.DataFrame(zip(searched_user_ids_list, 
                                        searched_tweet_text_list, 
                                        searched_tweet_createdAt_list, 
                                        searched_tweet_favoriteCount_list, 
                                        searched_tweet_retweetCount_list), 
                                    columns=["user_id", "text", "created_at", "favorite_count", "retweet_count"])
```
